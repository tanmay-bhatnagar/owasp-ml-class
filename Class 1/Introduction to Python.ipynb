```python
# classes and OOP
#NOT NECE
class MyClass:
    # x = 100
    def __init__(self, x=10):
        self.x = x
    
    def square(self):
        self.x = self.x**2
    
    def multiply(self, q):
        return self.x*q
    
    def mul_sq(self, q):
        self.x = self.multiply(q)
        self.square()
        return self.x

a = MyClass(120)
b = MyClass()

print a.x
print b.x
print a.mul_sq(10)
```
